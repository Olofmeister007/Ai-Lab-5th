```python
import random
class SimpleReflexAgent:
    def __init__(self, grid_width, grid_height):
        self.grid_width = grid_width
        self.grid_height = grid_height

        self.location = (random.randint(0, grid_width-1), random.randint(0,grid_height-1))

        # write code to initialize the agent at a random location
        self.move_right = True

    def perceive(self, location_status):
        x, y = self.location
        if location_status[x][y] == "dirty":
            return "dirty"
        else:
            return "clean"

    def decide(self, percept):
        x, y = self.location
        if percept == "dirty":
            return "suck"
        elif x == self.grid_width - 1 and y == self.grid_height - 1:
            return "done"
        elif self.move_right and y < self.grid_height - 1:
            return "move_right"
        elif self.move_right and y == self.grid_height - 1:
            self.move_right = False
            return "move_down"
        elif not self.move_right and y > 0:
            return "move_left"
        elif not self.move_right and y == 0:
            self.move_right = True
            return "move_down"


        #write code for condition where the robot might 'move up'

    def act(self, action):
        x, y = self.location
        if action == "suck":
            print(f"Location ({x}, {y}) is cleaned.")
        elif action == "move_right":
            print("Moving right.")
            self.location = (x, y + 1)
        elif action == "move_down":
            print("Moving down.")
            self.location = (x + 1, y)
        elif action == "move_left":
            print("Moving left.")
            self.location = (x, y - 1)

        #write code for 'move up' action

```

```python
def print_grid(grid):
    for row in grid:
        print(" ".join(row))
```

```python
def main():
    grid_width = int(input("Enter the width of the grid: "))
    grid_height = int(input("Enter the height of the grid: "))

    environment = [["clean" for _ in range(grid_height)] for _ in range(grid_width)]
    initialwidth=0
    finalwidth=grid_width-1
    initialheight=0
    finalheight=grid_height-1
    environment[random.randint(initialwidth, finalwidth)][random.randint(initialheight,finalheight)] = "dirty"
    environment[random.randint(initialwidth, finalwidth)][random.randint(initialheight,finalheight)] = "dirty"

    # write code to mark 2 cell as 'dirty' using random function

    agent = SimpleReflexAgent(grid_width, grid_height)

    while True:
        current_percept = agent.perceive(environment)
        current_action = agent.decide(current_percept)

        print(f"Current location: {agent.location} | Percept: {current_percept} | Action: {current_action}")

        if current_action == "done":
            print("Cleaning process complete.")
            break

        agent.act(current_action)
        x, y = agent.location
        environment[x][y] = "clean"

        print("Updated environment:")
        print_grid(environment)

if __name__ == "__main__":
    main()
```
